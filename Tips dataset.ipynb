```python
#import the necessary libraries
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

print('libraries loaded successfully')
```

```python
# load the tips data set
df = pd.read_csv(r'C:\Users\user\Downloads\tips.csv')
df.head()
```

Visualize the data

```python
plt.scatter(df['total_bill'], df['tip'])
plt.title('Total bill vs tip')
plt.xlabel('total(tb)')
plt.ylabel('tip(t)')
plt.show()
```

Choose features and target

```python
# independent variable
x = df['total_bill'].values # convert to numpy array
y = df['tip'].values # convert to numpy array

# linear regression
m, b = np.polyfit(x, y, 1) # 1 for linear fit
```

Train-Test_Split

```python
x_train, x_test, y_train, y_test = train_test_split(x.reshape(-1, 1), y, test_size=0.3, random_state=42)
```

Create and train the LinearRegression model

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

Evaluate the model

```python
y_pred = m* x_test + b
```

visualize the model

```python
# visualize total bill vs tip
plt.figure(figsize=(8, 5))
sns.scatterplot(x='total_bill',y='tip', data=df)
plt.plot(x_test, y_pred, color='red', label='Regression Line')
#Regression line
plt.title('Total bill vs Tip with Regression Line')
plt.xlabel('Total Bill (N)')
plt.ylabel('Tip (N)')
plt.legend()
plt.show()
```
